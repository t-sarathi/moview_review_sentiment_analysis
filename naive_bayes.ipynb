```python
# Text Cleaning:


# 1. sample 10000 rows
# 2. Remove HTML tags
# 3.Remove special characters
# 4.convert everything into lower case
# 5. removing stop words
# 6. stemming
```

```python
df=df.sample(10000)
```

```python
df.shape
```

```python
# checking the null values
df.isnull().sum()
```

```python
df['sentiment'].replace({'positive':1,'negative':0},inplace=True)
```

```python
df.sample(5)
```

```python
import re
```

```python
# function to clean the text:
def clean_text(text):
  clean_text = re.sub('<.*?>', '', text)
  return clean_text
```

```python
df['review']=df['review'].apply(clean_text)
```

```python
# convert into lower case :(it convert all the characters in to the lower case )
def lower_case(text):
  return text.lower()
```

```python
df['review']=df['review'].apply(lower_case)
```

```python
## Removing the special characters :
def remove_special_char(text):
 x=' '
 for i in text:
    if i.isalnum():
      x=x+i
    else:
      x=x+' '
 return x
```

```python
df['review']=df['review'].apply(remove_special_char)
```

```python
# remove the stops words:
import nltk
from nltk.corpus import stopwords

# Ensure stopwords are downloaded
nltk.download('stopwords')
stopwords.words('english')
```

```python
def  remove_stopwords(text):
  x=[]

  for i in text.split():

    if i not in stopwords.words('english'):
      x.append(i)
  y=x[:]
  x.clear()
  return y
```

```python
df['review']=df['review'].apply(remove_stopwords)
```

```python
df.head(5)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
y=[]

def stem_words(text):
  for i in  text:
    y.append (ps.stem(i))
  z=y[:]
  y.clear()
  return z
```

```python
df['review']=df['review'].apply(stem_words)
```

```python
df
```

```python
# join the key words back :
```

```python
def join_back(text):
  return " ".join(text)
```

```python
df['review']=df['review'].apply(join_back)
```

```python
df
```

```python
### Apply vectorization :
max_features_values = [1000, 2000, 5000, 10000, 20000]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
```

```python
X=cv.fit_transform(df['review']).toarray()
```

```python
X.shape
```

```python
Y=df['sentiment'].values  # X, Y are in list format
```

```python
Y
```

```python
# split the train and test data:
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

## Train the model :

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.metrics import accuracy_score,f1_score
```

```python
model1=GaussianNB()
model2=BernoulliNB()
model3=MultinomialNB()
```

```python
model1.fit(xtrain,ytrain)
model2.fit(xtrain,ytrain)
model3.fit(xtrain,ytrain)
```

```python
y_pred1=model1.predict(xtest)
y_pred2=model2.predict(xtest)
y_pred3=model3.predict(xtest)
```

```python
print("Gaussian",accuracy_score(ytest,y_pred1))
print("Bernoulli",accuracy_score(ytest,y_pred2))
print("Multinomial",accuracy_score(ytest,y_pred3))
```
